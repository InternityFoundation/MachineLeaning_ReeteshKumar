```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
df = make_blobs(n_samples=500,n_features=2,centers=5,random_state=42)
```

```python
df
```

```python
df[0].shape
```

```python
#Training data
df[0]
```

```python
#Outputs
df[1]
```

```python
plt.scatter(x=df[0][:,0],y=df[0][:,1],c=df[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(df[0])
```

```python
pred = kmeans.predict(df[0])
```

```python
pred
```

```python
#Either use predict or use labels_
```

```python
kmeans.labels_
```

```python
kmeans.cluster_centers_
```

```python
fig , (ax1,ax2)=plt.subplots(1,2,figsize=(10,7),sharey=True)

ax1.set_title('K Means')
ax1.scatter(x=df[0][:,0],y=df[0][:,1],c=pred,cmap='jet')

ax2.set_title('Original')
ax2.scatter(x=df[0][:,0],y=df[0][:,1],c=df[1],cmap='jet')
```

To find the best value of k try the same elbow method as discussed in KNN.
Note that KMeans is an unsupervised learning algorithm. Through make_blobs() method we already knew the clusters hence we were able to cross check but in real world data we won't be able to cross check like this.
