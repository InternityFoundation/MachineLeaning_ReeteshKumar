```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from apyori import apriori
```

```python
df = pd.read_csv('store_data.csv',header=None)
```

```python
df.shape
```

```python
df.head()
```

```python
data = df.values.astype('str').tolist()
```

```python
association_rules = apriori(data, min_support=0.0030, min_confidence=0.3, min_lift=4, min_length=2)
association_results = list(association_rules)
```

```python
association_results[0]
```

```python
for i in association_results:
    item=[]
    for j in i[0]:
        item.append(j)
    print("Rule:" ,item[0],item[1])
    print("Support:", i[1])
    print(i[2])
    #print("Confidence", i[2][2])
```
