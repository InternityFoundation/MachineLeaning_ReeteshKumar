```python
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
tips= sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.distplot(tips['total_bill'],kde=False,bins=10)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

```python
#CATEGORICAL PLOTS
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='day',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='sex',palette='rainbow')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',split=True)
```

```python
sns.swarmplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

```python
sns.factorplot(x='day',y='total_bill',data=tips,kind='violin')
```

```python
tp= tips.corr()
```

```python
sns.heatmap(tp,cmap='rainbow',linewidths=5,linecolor='black',annot=True)
```

```python
flights = sns.load_dataset("flights")
```

```python
flights.head()
```

```python
fp = flights.pivot_table(index='year',columns='month',values='passengers')
```

```python
sns.clustermap(fp,linewidths=5,linecolor='yellow',cmap='rainbow',standard_scale=1)
```

```python
iris= sns.load_dataset('iris')
```

```python
iris.head()
import matplotlib.pyplot as plt
```

```python
x= sns.PairGrid(data=iris)
x.map_diag(sns.kdeplot,color='red')
x.map_upper(plt.scatter)
x.map_lower(sns.kdeplot)
```
