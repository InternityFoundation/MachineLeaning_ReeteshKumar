```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.linspace(1,10,20)
```

```python
y=x**2
```

```python
plt.plot(x,y,'y-')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Title')
```

```python
plt.plot(x,y,'r--',label='1223')
plt.plot(y,x,'b--',label='12')
plt.legend(loc=0)
```

```python
# Object Oriented Method
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,1,1])

axes.plot(x,y,label='1st',color='green',marker='+',lw=3,ls='--',alpha=0.5,markersize=20,markerfacecolor="black",
          markeredgecolor='purple')
axes.plot(y,x,label='2nd',color='purple',marker='o',lw=4,ls=':',alpha=0.8,markersize=10,markerfacecolor="yellow",
         markeredgewidth=2,markeredgecolor='black')

axes.legend(loc=10)
```
