```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer['data']
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(df)
```

```python
sc_df = sc.transform(df)
```

```python
sc_df
```

```python
# PCA
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(sc_df)
```

```python
pca_df = pca.transform(sc_df)
```

```python
pca_df
```

```python
pca_df.shape
```

```python
sc_df.shape
```

```python
# Hence dimensionality reduction has taken place
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=pca_df[:,0],y=pca_df[:,1],c=cancer['target'])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
pca.components_
```

```python
df_components = pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_components
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_components)
```

```python
from sklearn.decomposition import KernelPCA
```

```python
kpca = KernelPCA(n_components=2,kernel='rbf')
kpca.fit(sc_df)
kpca_df= kpca.transform(sc_df)
```

```python
kpca_df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x=kpca_df[:,0],y=kpca_df[:,1],c=cancer['target'])
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

```python
#LDA
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(sc_df,cancer['target'])
lda_df = lda.transform(sc_df)
```
